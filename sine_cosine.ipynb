```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create an array of points for x-coordinate, starting from 0, incrementing by 0.1 and go till 9.4
x = np.arange(0, 3 * np.pi, 0.1)
```

```python
#create y-coordinate by plotting a sine wave and cos wave
sin_y = np.sin(x)
cos_y = np.cos(x)
```

```python
plt.title("Govindha Madhava")
plt.xlabel("my X data points")
plt.ylabel("my Y data points")

#plot a sin wave
plt.plot(x,sin_y)

#plot a cos wave
plt.plot(x,cos_y)


#give legegends
plt.legend(["Sine","CoSine"])

#show the graph
plt.show()
```
